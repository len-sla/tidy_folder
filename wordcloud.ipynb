# Word cloud
Word cloud is an effective way of visualizing the texts similary like charts for categorical data. From a pool of texts, you can see which words are the dominants. Just by looking at this visualization, you know the mode of the text. Additionally word cloud is displayed in custom shape characterised by image.

## 1. Importing libraries

```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import  WordCloud, STOPWORDS, ImageColorGenerator
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2. Presenting wordcloud for the book

```python
# reading the file ot string
with open("../GPT-2/A.txt" , 'r') as file:
    data = file.read().replace('\n', '')
    #book is about  Gut microbiobe, and inluece of food on our/microbiome healt
```

```python
len(data)
```

```python
stopwords =  set(STOPWORDS) # standard dictionary with stopwords
```

```python
# additional stop words
extra_w={"way", 'one',"help", "day", 'life', 'add', "find", 'without', 'body', "found", 'much', 'change', "new", "us", "human", "feel", "every", "start", "journal", "including"}
stopwords.update(extra_w)
```

Loading image which shape limits wordcloud

```python
mask= np.array(Image.open("../wordcloud/man.jpg"))
```

Creating wordcloud with given mask

```python
wordcl = WordCloud(stopwords=stopwords, background_color ="white", mask=mask, max_words=200, max_font_size=30, contour_width=.3,contour_color='white').generate(data)
fig= plt.figure(figsize=(20,30))
plt.imshow(wordcl, interpolation='bilinear')
plt.axis('off')

plt.show()
```

Checking other possibilities and parameters like colors min, max sizes of laterrs, saving to svg or jpg formats for later extrenal use

```python
wordcl = WordCloud(stopwords=stopwords, color_func=lambda *args, **kwargs:"white", mask=mask, max_words=200, max_font_size=40, contour_width=.3,contour_color='white').generate(data)
fig= plt.figure(figsize=(20,30))
plt.imshow(wordcl, interpolation='bilinear')
plt.axis('off')
plt.savefig("test3.svg")
plt.show()
```

```python
wordcl = WordCloud(stopwords=stopwords, color_func=lambda *args, **kwargs:"white", mask=mask, max_words=200, max_font_size=40,min_font_size=15,contour_width=.3,contour_color='white').generate(data)
fig= plt.figure(figsize=(20,30))
plt.imshow(wordcl, interpolation='bilinear')
plt.axis('off')
plt.savefig("glowa_min_max.jpg")
plt.show()
```
